```python
# Final working solution
# [AI Disclaimer example: Used ChatGPT with prompt “How to handle recursion base case” only for concept]

class Solution:
    def reverse(self, x,actual=0) -> int:
        if actual > (2**31)-1: # checks for positive overflow
            return 0
        if actual < -(2**31): # checks for negative overflow
            return 0
        if x == 0: # base case
            return actual
        if x < 0: # This is a one time call for negative numbers
            return -(self.reverse(-x,actual))
        return self.reverse(x // 10, actual * 10 + x % 10) # normal positive call


y = Solution()

print(y.reverse(10000))
print(y.reverse(-10000))
print(y.reverse(69))
print(y.reverse(4567))
print(y.reverse(12389))
print(y.reverse(-9876543211))
print(y.reverse(9999999999999999999999999))
print(y.reverse(-9999999999999999999999999))
```
