````python
# Final working solution
# [AI Disclaimer example: Used ChatGPT with prompt â€œHow to handle recursion base caseâ€ only for concept]
class Solution:
    def climbStairs(self, n, i=2, prev=1, curr=2) -> int:
        # n is number of steps
        # i is current increasing step
        # prev is i-2
        # curr is i-1
        # Like An = An-1 + An-2
        # Basecase
        if n <= 2:
            return n
        if i == n:
            return curr
        # Recursive call
        return self.climbStairs(n, i + 1, curr, prev + curr)
````
